```python
artist = 'Mobb Deep'
track_name = 'Survival of the Fittest'
```

```python
song = df[(df['artists'].str.contains(artist))& (df['track_name'] == track_name)]
if len(song) == 0:
    print("Song not found.")
song
```

```python
songvalues = song.iloc[0][features]
print(songvalues.shape)
print(songvalues.values)
songvalues
```

```python
similarity_scores = cosine_similarity(songvalues.values.reshape(1, -1), df[features])[0]
```
